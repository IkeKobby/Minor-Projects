## Dimensionality Reduction Techniques

### Principal Component Analysis for Exploratory Data Analysis

```python
## As required I shall perform a standardized the data using the standard scaler to have mean 0 and variance 1 for all variables
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(mfeat)
```

```python
### scaling the data
mfeat_scaled = scaler.transform(mfeat)
```

- As required the PCA library in python do not standardized the input data, hence the need to prior standardization.

```python
## Using PCA from the Sci-kit learn library
from sklearn.decomposition import PCA

### instantiating the PCA
pca = PCA(n_components=None,
    copy=True,
    whiten=False,
    svd_solver='auto',
    tol=0.0,
    iterated_power='auto',
    random_state=0,
)
```

```python
pca.fit(mfeat_scaled) ## fitting the scaled data into the PCA method.
```

```python
mfeat_pca = pca.transform(mfeat_scaled) ### transforming the scaled data into the principal components.
```

```python
### in this case after the PCA, I have 649 principal components to make selection using the elblow plot 
### for the best components
mfeat_scaled.shape, mfeat_pca.shape 
```

#### creating count array of length 649 to plot the inertia/variance explained by the PCA

```python
x = len(np.round(pca.explained_variance_ratio_))
x_ =np.linspace(1,649,649)
```

##### Scree plot for eigenvalues of the PCA results. 

```python
plt.scatter(x_, pca.explained_variance_)
```

##### Scree plot for the explained variance ratio of the PCA results. 

```python
plt.scatter(x_, pca.explained_variance_ratio_)
```

- The graph does not give explicit number of eigenvalue explanation but hence does depicts the consistency in its variance explained ratio. The scree plot does shows us 7 distinct eigen values before it turns into swarmpy constancy. Hence for the benefit of this work we shall consider at most 10 components as best to retain.

##### Cummulative plot for the explained variance ratio of the PCA results. 

```python
cum_variance = np.cumsum(pca.explained_variance_ratio_)
plt.plot(cum_variance)
```

### Manifold Spectral Embedding and Agglomerative Clustering

```python
from sklearn import manifold
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from time import time
```

```python
# 2D embedding of the mfeat dataset 
print("Computing embedding")
manifold_red = manifold.SpectralEmbedding()
manifold_mfeat = manifold_red.fit_transform(mfeat_scaled)
print("Done.")
```

```python
# The manifold spectral gave two components as the best representation of the entire dataset
manifold_mfeat.shape 
```

```python
plt.figure(figsize=(15,8))
plt.scatter(manifold_mfeat[:, 0], manifold_mfeat[:, 1], c =classes, cmap ='plasma')
plt.xlabel("First manifold component")
plt.ylabel("Second manifold component")
```

- From the manifold exploration, it's obvious that clusters are close to each other but in distinctive separation. Manifold seeks to give the various clusters that have close relation in some measure.

### Plotting the agglomerative Clustering with different linkage method

```python
def plot_clustering(X_reduced, labels, title=None):
    x_min, x_max = np.min(X_reduced, axis=0), np.max(X_reduced, axis=0)
    X_reduced = (X_reduced - x_min) / (x_max - x_min)

    plt.figure(figsize=(15, 8))
    for i in range(X_reduced.shape[0]):
        plt.text(X_reduced[i, 0], X_reduced[i, 1], str(classes.values[i]),
                 color=plt.cm.nipy_spectral(labels[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    plt.xticks([])
    plt.yticks([])
    if title is not None:
        plt.title(title, size=17)
    plt.axis('off')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
for linkage in ('ward', 'average', 'complete', 'single'):
    clustering = AgglomerativeClustering(linkage=linkage, n_clusters=10)
    t0 = time()
    clustering.fit(manifold_mfeat)
    print("%s :\t%.2fs" % (linkage, time() - t0))

    plot_clustering(manifold_mfeat, clustering.labels_, "%s linkage" % linkage)


plt.show()
```

- From the above plots for different linkage methods, it is obvious that the ward and complete methods give the best 10 clusters we can get from the dataset.

### Plotting Dendogram for the reduced data by the method of Manifold Spectral 

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)


X = manifold_mfeat

# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

model = model.fit(X)

plt.figure(figsize=(20,15))
plt.title('Hierarchical Clustering Dendrogram')

# plot the top five levels of the dendrogram
plot_dendrogram(model, truncate_mode='level', p=5)

plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```

- The dendrogram plot does not give ten clusters for the best cut that could be obtained for the top 5 level diagram. However, cutting to achieve the desired 10 classes also overfit the number of clusters required. This is but a fancy representation  of the reduced dataset from the manifold method.
